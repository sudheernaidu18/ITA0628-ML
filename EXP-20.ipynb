```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Sample monthly sales data
data = {
    "Month": [1, 2, 3, 4, 5, 6, 7, 8],
    "AdvertisingSpend": [2000, 2500, 3000, 3500, 4000, 4200, 4500, 4800],
    "Sales": [15000, 18000, 21000, 25000, 28000, 30000, 33000, 36000]
}

df = pd.DataFrame(data)
df
```

```python
X = df[["Month", "AdvertisingSpend"]]
y = df["Sales"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42
)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
# Predict sales for a future month
future_data = pd.DataFrame(
    [[9, 5000]],
    columns=["Month", "AdvertisingSpend"]
)

future_sales = model.predict(future_data)
print("Predicted Future Sales:", int(future_sales[0]))
```
